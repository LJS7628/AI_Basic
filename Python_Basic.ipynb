```python
# c = -1
# if 조건식 : c 변수의 값이 0보다 크다.
# elif 조건식 : c 변수의 값이 0보다 작다.
# else 조건식 : 없음 (c 변수의 값이 0이다)

c = -1
if c> 0:
  print('0보다 크다')
elif c< 0 :
  print('0보다 작다')
else :
  print('0이다')
```

```python
# for 반복문

num_list = [2,4,6]
for num in num_list :
  print(num)
```

```python
#for 반복문 : 리스트의 원소를 반복하여 2의 배수와 함께 출력한다.
for num in num_list :
  print(num*2)
```

```python
for num in [1,2,3]:
  print("기존: ",num)
  print("2배: ",num*2)
  print("\n")
```

```python
for num in [1,2,3]:
  print(f"기존:{num}, 2배:{num*2}")
```

```python
# for 반복문 ; 리스트의 원소를 반복해서 2의 배수를 계산하고,
# 빈 리스트(double)에 2의 배수를 원소로 추가한다. append

double = list()
for num in [1,2,3] :
  double.append(num*2)

double
```

```python
#list comprehension

[x*2 for x in [1,2,3]]
```
